## **Problem 1**
##### Using exponent

```python
# Constant for thickness of paper
THICKNESS_OF_PAPER = 0.00008

number_of_times_folded = int(input("How many times is the paper folded?: "))
folded_thickness = THICKNESS_OF_PAPER * (2**number_of_times_folded)

print("The thickess of the paper after beeing folded {} times is {} meters".format(number_of_times_folded, folded_thickness))
```

## **Problem 2**
##### Converting to Kilometers

```python
print("The thickess of the paper after beeing folded {} times is {: .2f} kilometers".format(number_of_times_folded, (folded_thickness/1000)))
```
