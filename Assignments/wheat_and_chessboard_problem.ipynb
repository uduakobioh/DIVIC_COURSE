## Problem 1
**2x2 square**

```python
import numpy as np
n_rows = 2
n_columns = 2
two_by_two_list = [1]
for i in range((n_rows*n_columns)-1):
    two_by_two_list.append(2*two_by_two_list[-1])
print("Arrange wheat on a plate of 2x2 squares (list)：{}".format(two_by_two_list))

two_by_two_board_ndarray = np.array(two_by_two_list).reshape(n_rows, n_columns)
print("Arrange wheat on a plate of 2x2 squares (ndarray)：{}".format(two_by_two_board_ndarray))
```

## Problem 2
**function to determine nxm array**

```python
def determine_array(n_rows, n_columns):
    n_by_m_list = [1]
    for i in range((n_rows*n_columns)-1):
        n_by_m_list.append(2*n_by_m_list[-1])
    n_by_m_ndarray = np.array(n_by_m_list).reshape(n_rows, n_columns)
    return n_by_m_ndarray
    
eight_by_eight_ndarray = determine_array(8, 8)
print("Arrange wheat on a plate of 8x8 squares (ndarray)：{}".format(eight_by_eight_ndarray))
```

## Problem 6
**Expansion to nxm mass by Another Calculation Method**

```python
def using_np_append(n_rows, n_columns):
    n_squares = n_rows*n_columns
    board_ndarray = np.array([1]).astype(np.uint64)
    for _ in range(n_squares - 1):
        board_ndarray = np.append(board_ndarray, 2*board_ndarray[-1])
    board_ndarray = board_ndarray.reshape(n_rows, n_columns)
    return board_ndarray
np_append_array = using_np_append(8,8)
print("Arrange wheat on a plate of 4 squares (ndarray)：{}".format(np_append_array))
```

```python
def using_broadcast(n_rows, n_columns):
    n_squares = n_rows*n_columns
    indices_of_squares = np.arange(n_squares).astype(np.uint64).reshape(n_rows, n_columns)
    board_ndarray = 2**indices_of_squares
    return board_ndarray

broadcast_array = using_broadcast(8,8)
print("Arrange wheat on a plate of 8x8 squares (ndarray)：{}".format(broadcast_array))
```
