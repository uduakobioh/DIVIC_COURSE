## Problem 4
**Examining the Iris Dataset**

The Iris flower data set or Fisher's Iris data set is a multivariate data set used and made famous by the British statistician and biologist Ronald Fisher in his 1936 paper It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. Two of the three species were collected in the Gaspé Peninsula "all from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus".

Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other. Fisher's paper was published in the Annals of Eugenics (today the Annals of Human Genetics) and includes discussion of the contained techniques' applications to the field of phrenology.

The dataset contains a set of 150 records, 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor) under five attributes: sepal length, sepal width, petal length, petal width and species. The Iris Setosa is linearly separable from the other but the other two are not linearly separable from each other. The iris data set is widely used as a beginner's dataset for machine learning purposes. The dataset is included in R base and Python in the machine learning library scikit-learn, so that users can access it without having to find a source for it.

Difference Between Box Plots & Violin Plot & Advantages of Each

### Box Plot
  
A box plot, also known as a box and whisker plot, is a standardized way ofdisplaying the distribution of data based on five summary statistics: minimum,first quartile (Q1), median, third quartile (Q3), and maximum. In a box plot, arectangular box is drawn to represent the interquartile range (IQR), which isthe range between the first and third quartiles. A line inside the box representsthe median value. Lines called whiskers are drawn from the box to theminimum and maximum values, or a certain distance from the box, usually 1.5times the IQR.  
  
<br>

### Violin Plot
  
A violin plot, on the other hand, is a combination of a box plot and a kerneldensity plot. It displays a kernel density estimation of the underlyingdistribution of the data along with the box plot. The width of the violin at aparticular point represents the density or frequency of the data at that point.The kernel density plot is symmetrical along the vertical axis, with the medianvalue represented by a horizontal line inside the violin. The two halves of theviolin plot mirror each other, indicating the symmetry of the data.  
<br>
***
<br>
The main difference between the two plots lies in the way they represent the distribution of the data. A box plot gives a summary of the five-number summary of the data while ignoring the distributional shape. It is more usefulwhen comparing the distribution of data between different groups. Box plots are generally morecompact and easier to read, particularly when comparing multiple datasetsside by side. They also make it easy to identify outliers, which can be useful incertain contexts.  
   
Violin plots, on the other hand, are useful when you want to see more detailabout the distribution of the data. A violin plot shows the shape of the distribution of the data, alongwith the summary statistics. It is useful when we want to see the full picture ofthe data distribution, particularly when it is non-normal, and also compare distributions between groups. However, violin plots can be more difficult to read than boxplots, particularly when the data is very skewed or when there are multiple peaks.  
<br>
***
<br>

#### Advantages of Box Plots
-Organises large amounts of data<br>
-Visualises outliers<br>
-Provides a clear summary  
  
<br>

#### Advantages of Violin Plots
-Shows the entire distribution of the data<br>
-Identify unusual groups of data<br>
