## **Problem 1**
##### Number of Rice Grains on 100th Day

```python
number_of_days = 100
grains_per_day = 1
total_grains = 1
list_of_grains_per_day = [grains_per_day]
list_of_total_grains_at_the_end_of_every_day = [total_grains]
for i in range(number_of_days-1):
    grains_per_day *= 2
    total_grains += grains_per_day
    list_of_grains_per_day.append(grains_per_day)
    list_of_total_grains_at_the_end_of_every_day.append(total_grains)
print(f"The number of rice grains received on day {number_of_days} is {grains_per_day} and the total number of rice grains after {number_of_days} days is {total_grains}")
```

#### Checking the number of entries in `list_of_total_grains_at_the_end_of_every_day`

**Creating the Line Chart to Show the Number of Grains Received per Day**

```python
import matplotlib.pyplot as plt
plt.title(label="Number of Rice Grains per day")
plt.ylabel(ylabel="Number of Rice Grains Received")
plt.xlabel(xlabel="Number of Days")
plt.plot(list_of_grains_per_day)
```

**Line Chart to Show the Total Grains Received at the End of Each Day**

```python
plt.title(label="Total RIce Grains at the End of Each Day")
plt.ylabel(ylabel="Total Rice Grains")
plt.xlabel(xlabel="Number of Days")
plt.plot(list_of_total_grains_at_the_end_of_every_day)
```

## **Problem 2**
##### Number of Rice Grains outside the 100th day

```python
def compute_sorori_shinzaemon(n_days=100):
    """A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameteres
    -----------
    n_days: int
        Number of days to get rice (default : 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice you get in one day (list length equals n_days)
    list_total_grains : list
        List of total number of rice received by a certain day (list length is equal to n_days)
    """

    # Write the code here.
    total_grains_received = 1
    grains_received_per_day = 1
    list_n_grains = [grains_received_per_day]
    list_total_grains = [total_grains_received]
    for i in range(n_days-1):
        grains_received_per_day *= 2
        total_grains_received += grains_received_per_day
        list_n_grains.append(grains_received_per_day)
        list_total_grains.append(total_grains_received)

    return list_n_grains, list_total_grains

list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=365)
```

## **Problem 3**
##### How many People Can Live for How Many Days with The Rice We Get

```python
def number_of_days_people_in_class_can_live_with_total_grains_received(number_of_days, number_of_people):
    # Constants to define how much grains of rice a person eats per day
    NUMBER_OF_GRAINS_IN_ONE_SERVING_PER_PERSON = 3000
    NUMBER_OF_SERVINGS_EATEN_PER_DAY_PER_PERSON = 3
    TOTAL_GRAINS_EATEN_PER_DAY = NUMBER_OF_GRAINS_IN_ONE_SERVING_PER_PERSON * NUMBER_OF_SERVINGS_EATEN_PER_DAY_PER_PERSON * number_of_people
    
    total_grains_received = 1
    grains_received_per_day = 1
    for i in range(number_of_days-1):
        grains_received_per_day *= 2
        total_grains_received += grains_received_per_day
        
    number_of_days_grains_received_will_last = total_grains_received // TOTAL_GRAINS_EATEN_PER_DAY
    return number_of_days_grains_received_will_last

number_of_days = number_of_days_people_in_class_can_live_with_total_grains_received(365, 50)
print(f"The total number of grains received will last everyone in this class for {number_of_days} days")
```
