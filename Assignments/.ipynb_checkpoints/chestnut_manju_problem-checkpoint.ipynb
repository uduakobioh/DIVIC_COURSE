##### Calculating How long it Will Take for Chestnut buns to Cover the Whole Solar System

```python
# Volume of constant values in cubic centimeters
SOLAR_SYSTEM_VOLUME = 1.931e70
CHESTNUT_BUN_VOLUME = 15

total_volume = CHESTNUT_BUN_VOLUME
count = 0 # Number of times the volume has been doubled
while (total_volume <= SOLAR_SYSTEM_VOLUME):
    total_volume *= 2
    # print(total_volume)
    count += 1
    #cprint(count)

# Finding total time taken if volume doubles every 5 minutes
total_time_minutes = count * 5
total_time_hours = total_time_minutes/60

print(f"It will take {total_time_hours} hours for chestnut buns to completely fill the solar system")
```

##### Function toCalculate how long it Takes and Object sprinkled with Vine to Fill Up a Space

```python
def calculate_time_to_fill_space(volume_of_object, volume_of_space):
    """
    Function calculates how long it will take any object sprinkled with belvine to fill up an arbitrary space
    
    Parameters:
    volume_of_object: (int/float) the volume of the object sprinkled with belvine
    volume_of_space: (int/float) the volume of the space to be filed up
    
    Returns:
    time_hours: (float) the total time taken in hours
    list_volumes: (list) a list of the volume filled up over time"""
    count = 0
    total_volume = volume_of_object
    list_volumes = [total_volume]
    
    while (total_volume <= volume_of_space):
        total_volume *= 2
        count += 1
        list_volumes.append(total_volume)
        
    time_minutes = count * 5
    time_hours = time_minutes/60
    
    return time_hours, list_volumes
```

```python
chestnut_fill_up_time, chestnut_volumes_per_time = calculate_time_to_fill_space (CHESTNUT_BUN_VOLUME, SOLAR_SYSTEM_VOLUME)
print(f"It will take {chestnut_fill_up_time} hours to fill up the solar system with chestnut buns.")
```

##### Graph to show the volume of the solar system filled up by chestnut buns over time

```python
import matplotlib.pyplot as plt
plt.title(label="Volume of Solar System Filled Up by Chestnut Buns Over time")
plt.ylabel(ylabel="Volume of Space Filled Up")
plt.xlabel("Time Elapsed in Minutes")
plt.plot(range(0,len(chestnut_volumes_per_time*5),5), chestnut_volumes_per_time)
```
