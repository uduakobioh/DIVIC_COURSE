## Problem 4
**Dealing with Missing Values**

Check for Missing Values for Each Feature

```python
import missingno as msno
%matplotlib inline
msno.matrix(df_train.sample(250), labels=True, fontsize=10)
```

Checking the percentage of missing values

```python
total = df_train.isnull().sum()
percent = (df_train.isnull().sum())/(df_train.isnull().count())
missing_values_df = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_values_df[missing_values_df.Total > 0]
```

Delete Feautres with 5 or more Missing Values

```python
drop_columns = missing_values_df[missing_values_df["Total"] >= 5].index
df_train.drop(drop_columns, axis=1, inplace=True)
```

Deleting Rows with Missing Values

```python
df_train = df_train.dropna()
```

Comfirming that there are no missing values in any row/column

```python
df_train.info()
```

## Problem 5
**Researching Terminology**

### **SKWEWNESS**
Skewness is a measure of symmetry or asymmetry of a data distribution. It measures the extent to which the data is shifted towards one side, away from the symmetrical normal distribution.  
Positive skewness indicates a longer tail on the right side of the distribution, while negative skewness indicates a longer tail on the left side.  
If the values of a specific feature are skewed, depending on the model, skewness may violate model assumptions or may reduce the interpretation of feature importance.
<br>
##### **TYPES OF SKEWNESS**  
_Positive Skewness or Right-Skewness_:  
In statistics, a positively skewed distribution has a long right tail. It is a sort of distribution where the measures are dispersing, unlike symmetrically distributed data where all measures of the central tendency (mean, mode, and median) equal each other. This makes Positively Skewed Distribution a type of distribution where the mean, median and mode of the distribution are positive rather than negative or zero. 
In positively skewed, the mean of the data is greater than the median(a large numver of data pushed on the right side). In other words, the results are bent towards the lower side. The mean will be more than the median as the median is the middle value and the mode is always the most frequent value.  
Extreme positive skewnees is not desirable for a distribution, as a high level of skewness can cause misleading results. For positive skewness, the most famous transformation to shift the distribution closer to the normal distribution is the log transformation. The log transformation proposes the calculationsof the natural logarithm for each value in the dataset.  
_Negative Skewness or Left-Skewness_:  
A negatively skewed or left-skewed distribution has a long left tail. It is the complete opposite of a positively skewed distribution. It refers to a type of distribution that has most of the values on the right side, which causes the a stretched out left tail. Here, the mean of the data is less than the median. The mean, median and mode are negative rather than zero.

### **KURTOSIS**
Kurtosis measures the shape of a probability distribution. It gives information about the tails and peakedness of the distribution compared to a normal distribution. Positive kurtosis means the distribution is more peaked with heavier tails. Whereas, negative kurtosis suggests the distribution is flatter with a lighter tail. Kurtosis is useful in determining the presence of outliers. Distributions with high kurtosis have a heavy tail which indicates the presence of outliers in the distribution. The opposite is the case for distributions with low kurtosis.

## Problem 6
**Confirming Distribution**

```python
sns.distplot(df_train['SalePrice'])
```

```python
df_train['SalePrice'].skew()
```

```python
df_train['SalePrice'].kurt()
```

The distribution of the `SalePrice` column is shifted to the right and deviates from the normal distribution. It has positive skewness. It also has a high kurtosis with a value of 6.53. This indicates tha data is heavy tailed and contains outliers that will need to be dealt with.

Performing a Logarithmic Transformation

```python
df_train['SalePriceLog'] = np.log(df_train['SalePrice'])
df_train.drop('SalePrice', axis=1, inplace=True)
```

```python
sns.distplot(df_train['SalePriceLog'])
```

```python
df_train['SalePriceLog'].skew()
```

```python
df_train['SalePriceLog'].kurt()
```

Originally, the distribution of the variable `SalePrice` deviates from the normal distribution and is positively skewed. By applying a logarithmic transformation, the skewness is corrected. This allows the distribution to fit the normal distribution. The kurtosis has also been brought down to almost zero. This means the transformation makes the distribution more light tailed and now lacks outliers.
