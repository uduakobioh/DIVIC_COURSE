## Problem 1
**Creating Random Number**

```python
import numpy as np
```

```python
mean_array = [-3,0]
cov_matrix = [[1.0, 0.8], [0.8, 1.0]]
```

```python
mean_array
```

```python
cov_matrix
```

```python
np.random.seed(0)
x = np.random.multivariate_normal(mean=mean_array, cov=cov_matrix, size=(500))
```

```python
x
```

```python
x.shape
```

## Problem 2
**Visualisation Using Scatter Plot**

```python
import matplotlib.pyplot as plt
x1 = x[:,0]
x2 = x[:,1]
plt.title("Visualisation Using Scatter Plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x1,x2)
plt.show()
```

## Problem 3
**Visulaisation Using Histogram**

```python
plt.title("Histogram of x1")
plt.xlabel("x1")
plt.ylabel("Frequency")
plt.xlim(right=3)
plt.xlim(left=-6)
plt.hist(x1)
plt.show()
```

```python
plt.title("Histogram of x2")
plt.xlabel("x2")
plt.ylabel("Frequency")
plt.xlim(right=3)
plt.xlim(left=-6)
plt.hist(x2)
plt.show()
```

## Problem 4
**Addition of Data**

```python
y_mean = [0, -3]
y_cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
y = np.random.multivariate_normal(mean=y_mean, cov=y_cov, size=500)
y
```

```python
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
line1 = plt.scatter(x1, x2)
line2 = plt.scatter(y[:, 0], y[:,1])
plt.legend((line1, line2), (0,1))
plt.show()
```

## Problem 5
**Data COmbination**

```python
final_data = np.vstack((x, y))
final_data
```

```python
final_data.shape
```

## Problem 6
**Labelling**

```python
zeros = np.zeros([500,1])
ones = np.ones([500,1])
labels = np.vstack((zeros, ones))
labels
```

```python
labelled_data = np.hstack((final_data, labels))
labelled_data
```
